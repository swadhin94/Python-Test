```python
def process_numbers(numbers, choice):
    # Check choice validity
    if choice not in ["even", "odd"]:
        return -1

    even_numbers = [num for num in numbers if num % 2 == 0]
    odd_numbers = [num for num in numbers if num % 2 != 0]

    # Prepare the result dictionary
    result_data = {
        "even_numbers": even_numbers,
        "odd_numbers": odd_numbers,
        "even_count": len(even_numbers),
        "odd_count": len(odd_numbers)
    }

    # Filter numbers based on choice and return the relevant data
    if choice == "even":
        return {"filtered": result_data["even_numbers"], "even_count": result_data["even_count"], "odd_count": result_data["odd_count"]}
    else: # choice == "odd"
        return {"filtered": result_data["odd_numbers"], "even_count": result_data["even_count"], "odd_count": result_data["odd_count"]}


# ---- Main Program ----
while True:
    user_input = input("Enter numbers separated by space: ").strip()

    # Flag to track if an error occurred in this iteration
    error_occurred = False

    # Check if input is empty
    if not user_input:
        print("Error: You have to enter at least 1 number")
        error_occurred = True
    else:
        parts = user_input.split()
        numbers = []
        valid_numbers_input = True # Flag for number parsing validity

        for p in parts:
            if not p.isdigit():
                print("Error: All values must be integers and \u2265 0")
                valid_numbers_input = False
                error_occurred = True
                break
            numbers.append(int(p))

        if valid_numbers_input: # Only proceed if numbers input was valid
            choice = input("Enter 'even' or 'odd': ").lower()

            result_info = process_numbers(numbers, choice)

            if result_info == -1:
                print("Error: Choice must be 'even' or 'odd'")
                error_occurred = True
            else:
                print("Result:", result_info["filtered"])
                print(f"Count of Even Numbers: {result_info['even_count']}")
                print(f"Count of Odd Numbers: {result_info['odd_count']}")

    # After processing (either error or success), ask the user if they want to continue
    if error_occurred:
        continue_input = input("An error occurred. Do you want to try again? (yes/no): ").lower()
    else:
        continue_input = input("Operation completed. Do you want to perform another operation? (yes/no): ").lower()

    if continue_input != 'yes':
        print("Exiting program.")
        break
```
